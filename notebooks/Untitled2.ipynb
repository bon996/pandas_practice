```python
import pandas as pd

df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando","Baltimore"],
    "temperature": [21,14,35,32],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","DC"],
    "humidity": [65,68,75],
})
df2
```

```python
df3=pd.merge(df1,df2,on="city",how="right",indicator=True)
df3
```

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35, 38],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "humidity": [65,68,71],
})
df2
```

```python
df3=pd.merge(df1,df2,on="city",how="outer")
df3
```

```python
df1 = pd.DataFrame({ 
    "city" : ["newyork", "chicago","orlando","baltimore"],
    "temperature" : [21,14,35,38],
    "humidity" : [65,68,71,75]
})
df1
```

```python
df2 = pd.DataFrame({
    "city" : ["chicago","newyork","san diago"],
    "temperature" : [21,14,35],
    "humidity" : [65,68,71]
})
df2
```

```python
df3 = pd.merge(df1,df2,on="city", suffixes=('_left','_right'))
df3
```
