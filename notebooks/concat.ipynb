```python
import pandas as pd

india_weather =pd.DataFrame({
    "city" : ["mumbai","delhi","banglore"],
    "temperature" : [32,45,30],
    "humidity" : [80,60,78]
})
india_weather
```

```python
usa_weather =pd.DataFrame({
    "city" : ["NewYork","chicago","orlando"],
    "temperature" : [21,14,35],
    "humidity" : [68,65,75]
})
usa_weather
```

```python
pd.concat((india_weather,usa_weather))
```

```python
pd.concat((india_weather,usa_weather), ignore_index=True)
```

```python
df = pd.concat((india_weather,usa_weather), keys=["india" , "usa"])
df
```

```python
df.loc["india"]
```

```python
df.loc["usa"]
```

```python
temperature_df = pd.DataFrame({
    "city" : ["NewYork","chicago","orlando"],
    "temperature" : [21,14,35]
}, index=[0,1,2])
temperature_df
```

```python
windspeed_df = pd.DataFrame({
    "city" : ["orlando","NewYork"],
    "windspeed" : [7,12]
}, index=[2,0])
windspeed_df
```

```python
df = pd.concat([temperature_df ,windspeed_df], axis=1)
df
```
